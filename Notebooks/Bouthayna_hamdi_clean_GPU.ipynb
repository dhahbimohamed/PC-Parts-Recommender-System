```python
import pandas as pd
import numpy as np
```

```python
prices=pd.read_csv('/content/pangoly_gpu_prices.csv')
```

```python
prices.info()
```

```python
specs=pd.read_csv('/content/pangoly_gpu_specifications.csv')
```

```python
specs.info()
```

```python
remove_specs_cols=['Page','Product_ID','List_Price','Availability','Reviews','Rating','Price','Release Date']
```

```python
new_specs=specs.drop(remove_specs_cols,axis=1)
```

```python
new_specs.head()
```

```python
new_specs=new_specs.rename(columns={'URL':'Product_URL'})
```

```python
new_specs.info()
```

```python
new_specs.columns=new_specs.columns.str.replace(' ','_')
```

```python
new_specs.info()
```

```python
remove_prices_cols=['Product_Name','Seller','Availability','Shipping','Currency']
```

```python
new_prices=prices.drop(remove_prices_cols,axis=1)
```

```python
new_prices.info()
```

```python
merged=pd.merge(new_specs,new_prices,on='Product_URL',how='outer')
```

```python
merged.info()
```

```python
merged.drop('Product_URL',axis=1,inplace=True)
```

```python
merged.head()
```

```python
merged.drop('Brand',axis=1,inplace=True)
```

```python
merged['Length'].head()
```

```python
merged['TDP'].head()
```

```python
for col in ['Length','Power_Connectors','TDP','Boost_Clock','Base_Clock','Memory_Size']:
  merged[col]=merged[col].str.replace(r'[^\d.]','',regex=True)
  merged[col]=pd.to_numeric(merged[col],errors='coerce')
merged.info()
```

```python
merged.drop('GPU_Series',axis=1,inplace=True)
```

```python
merged.isnull().sum()
```

```python
numerical_cols=['Length','Power_Connectors','TDP','Boost_Clock','Base_Clock','Memory_Size','HDMI','Expansion_Slots','Price']
merged.dropna(subset=numerical_cols,inplace=True)
merged.isnull().sum()
```

```python
merged.shape
```

```python
merged.drop_duplicates(inplace=True)
```

```python
merged.shape
```

```python
merged.to_csv("Cleaned_GPU_Data.csv",index=False)
```
