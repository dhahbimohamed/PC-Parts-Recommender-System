```python
import pandas as pd
import numpy as np
```

```python
prices=pd.read_csv('/content/pangoly_motherboard_prices.csv')
```

```python
prices.info()
```

```python
remove_cols_prices=['Shipping','Currency','Seller','Product_Name']
```

```python
new_prices=prices.drop(remove_cols_prices,axis=1)
```

```python
new_prices.rename(columns={'Product_URL':'URL'},inplace=True)
```

```python
new_prices.info()
```

```python
specs=pd.read_csv('/content/pangoly_motherboard_specifications.csv')
```

```python
specs.info()
```

```python
remove_cols_specs=['Page','Product_ID','List_Price','Availability','Reviews','Rating','Price','Release Date']
```

```python
new_specs=specs.drop(remove_cols_specs,axis=1)
```

```python
new_specs.info()
```

```python
merged=pd.merge(new_prices,new_specs,on='URL',how='outer')
```

```python
merged.info()
```

```python
merged.head()
```

```python
for col in ['Max RAM', 'Wi-Fi Generation', 'Audio Channels']:
    merged[col] = merged[col].str.replace(r'[^\d\.]', '', regex=True)
    merged[col] = pd.to_numeric(merged[col], errors='coerce')
merged.info()
```

```python
merged.isnull().sum()
```

```python
drop_cols=['USB 3.2 Gen2x2 Headers','LAN Chipset','Audio Chipset','Memory Type']
```

```python
merged.drop(columns=drop_cols,inplace=True)
```

```python
merged.info()
```

```python
numeric_cols = ['Price', 'RAM Slots', 'Max RAM', 'PCIe x16 Slots', 'PCIe x4 Slots', 'PCIe x1 Slots','M.2 Ports','SATA Ports','SATA Express Ports','Audio Channels','Wi-Fi Generation','USB 2.0 Headers','USB 3.2 Gen1 Headers','USB 3.2 Gen2 Headers']
merged.dropna(subset=numeric_cols, inplace=True)

merged.isnull().sum()
```

```python
merged.drop_duplicates()
```

```python
merged.shape
```

```python
merged.to_csv('clean_motherbords_Data',index=False)
```
