```python
import pandas as pd
import numpy as np
```

```python
prices=pd.read_csv('/content/pangoly_ram_prices.csv')
```

```python
prices.info()
```

```python
remove_cols_price=['Currency','Shipping','Seller','Availability','Product_Name']
```

```python
new_prices=prices.drop(remove_cols_price,axis=1)
```

```python
new_prices.rename(columns={'Product_URL':'URL'},inplace=True)
```

```python
new_prices.info()
```

```python
specs=pd.read_csv('/content/pangoly_ram_specifications.csv')
```

```python
specs.info()
```

```python
remove_cols_specs=['Page','Product_ID','List_Price','Reviews','Rating','Price','Release Date','Model']
```

```python
new_specs=specs.drop(remove_cols_specs,axis=1)
```

```python
merged=pd.merge(new_prices,new_specs,on='URL',how='outer')
```

```python
merged.info()
```

```python
merged.head()
```

```python
for col in ['Speed', 'Size', 'Voltage','First Word Latency']:
    merged[col] = merged[col].str.replace(r'[^\d.]', '', regex=True)
    merged[col] = pd.to_numeric(merged[col], errors='coerce')
merged.info()
```

```python
merged.isnull().sum()
```

```python
merged.tail()
```

```python
null_cols=['Color','ECC Support','Heat Spreader','Timing']
```

```python
merged=merged.drop(null_cols,axis=1)
```

```python
merged.info()
```

```python
numer_cols=['Price','Size','Speed','CAS Latency','First Word Latency','Voltage']
```

```python
merged.dropna(subset=numer_cols,inplace=True)
```

```python
merged.isnull().sum()
```

```python
merged.shape
```

```python
merged.to_csv('clean_data_RAM')
```
