```python
import pandas as pd
import numpy as np
```

```python
specs = pd.read_csv('/content/pangoly_cpu_specifications.csv')
prices = pd.read_csv('/content/pangoly_cpu_prices.csv')
```

```python
specs.info()
```

```python
remove_cols_specs = ['Release Date','Price','Rating','List_Price','Product_ID','Page','Reviews','Availability','Series']
```

```python
new_specs = specs.drop(remove_cols_specs,axis=1)
```

```python
new_specs.head()
```

```python
prices.info()
```

```python
remove_cols_prices = ['Shipping','Currency']
new_prices = prices.drop(remove_cols_prices,axis=1)
```

```python
new_prices.head()
```

```python
new_prices = new_prices.rename(columns={"Product_URL": "URL"})
```

```python
merged = pd.merge(new_specs, new_prices, on='URL', how='outer')
merged.shape
```

```python
merged = merged.drop(columns=['Product_Name', 'URL'])
merged.head()
```

```python
merged.dropna(subset=['Price'], inplace=True)
merged.reset_index(drop=True, inplace=True)
merged.shape
```

```python
merged.info()
```

```python
for col in ['Base Clock', 'Turbo Clock', 'TDP', 'L3 Cache']:
    merged[col] = merged[col].str.replace(r'[^\d\.]', '', regex=True)
    merged[col] = pd.to_numeric(merged[col], errors='coerce')
merged.info()
```

```python
merged.isnull().sum()
```

```python
numeric_cols = ['Turbo Clock', 'Base Clock', 'TDP', 'L3 Cache', 'Cores', 'Threads']
merged.dropna(subset=numeric_cols, inplace=True)

merged.isnull().sum()
```

```python
merged.head()
```

```python
merged.to_csv('cleaned_cpus.csv', index=False)
```
